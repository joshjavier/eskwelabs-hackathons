# House Star Wars

    A music production company that would like to identify past successful releases/ flops and identify song qualities that the streaming market favorably responds to.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('merged_chart_tracks.csv')
```

```python
df.head()
```

Filter sucessful tracks according to criteria set.

```python
df2 = df[(df['position']<=50)&(df['date']>'2019-04-04')].groupby(['track_id','track_name','artist'])['position'].count().sort_values(ascending=False).reset_index()
df2=df2[df2['position']>=63]
df2
```

Filter flop tracks according to criteria set.

```python
df = pd.read_csv('merged_chart_tracks.csv')
```

```python
df1 = df[(df['position']>50)&(df['date']>'2019-04-04')].groupby(['track_id','track_name','artist'])['position'].count().sort_values(ascending=False).reset_index()
df1=df1[df1['position']==1]
```

```python
df1
```

```python
track_df = pd.read_csv('spotify_track_data_top200_ph.csv')
```

```python
success_df = track_df[track_df['track_id'].isin(df2['track_id'])]
flop_df = track_df[track_df['track_id'].isin(df1['track_id'])]
success_df = success_df.drop(columns = ['Unnamed: 0','index'])
flop_df = flop_df.drop(columns = ['Unnamed: 0','index'])
```

Compare success & flop statistics:

```python
success_df.describe()
```

```python
flop_df.describe()
```

Add success column & position column to original track_df.

Success: 1 if in the success_df, 0 if not

Position: Count of how many time success tracks are in top 50, otherwise 0

```python
track_df['success'] = track_df['track_id'].isin(df2['track_id'].values).apply(lambda x: int(x))

track_df
```

```python
df = pd.merge(track_df, df2, how='left').fillna(0)
df
```

Look at correlation data:

```python
corr = df.corr()
corr.style.background_gradient(cmap = 'coolwarm')
```

No correlation stood out for success.
But we pick out metrics with highest correlation.
1. Energy: Success is negatively correlated.
2. Acousticness: Success is positively correlated.


```python
df_filtered = pd.merge(success_df, flop_df, how='outer')
```
